```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Microsoft_Stock.csv")  # Adjust the filename/path as necessary
features = data.drop(columns=["Date", "Volume"])  # Dropping non-numeric and non-feature columns

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Perform PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(features_scaled)

# Create a DataFrame for PCA results
pca_df = pd.DataFrame(pca_transformed, columns=['Principal Component 1', 'Principal Component 2'])

# Dummy industry sector grouping (Replace with actual industry sector data)
pca_df["Sector"] = np.random.choice(["Tech", "Finance", "Energy"], size=len(pca_df))  # Simulated data

# Plot the PCA results
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='Sector',
    palette='viridis',
    data=pca_df
)
plt.title('PCA Scatter Plot (Grouped by Sector)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid()
plt.show()
```
