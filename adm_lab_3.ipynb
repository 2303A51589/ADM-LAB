```python

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Labels
```

```python
X
```

```python
y
```

```python
# Step 2: Preprocess the data (standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 3: Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
# Step 4: Visualize the reduced features (2D scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k',
s=100)
plt.title('PCA of Iris Dataset (2 Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target Class')
plt.show()
```

```python
print("Explained Variance Ratio of Each Component:")
print(pca.explained_variance_ratio_)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns  # Import Seaborn

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Select features for PCA (e.g., 'age', 'fare', 'pclass')
X = titanic[['age', 'fare', 'pclass']]
X = X.dropna()  # Remove rows with missing values
y = titanic.survived[X.index]

# Preprocess the data (standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce to 3 dimensions
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Create the figure and axes
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with corrected dimensions
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap='viridis', edgecolor='k', s=100)

# Set labels and title
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA of Titanic Dataset (3 Components)')

# Display the plot
plt.show()
```
